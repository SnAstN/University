## Разбор всех столбцов таблицы

```python
def Remove(df, C):
    columns = C
    result = df
    #  позволяет удалять из DataFrame строки с пропущенными значениями
    initial_shape = df.shape[0]
    df = df.dropna(subset=columns)
    final_shape = df.shape[0]
    if final_shape == initial_shape:
        print("Пропущенных значений нет")
    else:
        removed_rows = initial_shape - final_shape 
        print(f"Есть пропущенные значения. Удалено строк: {removed_rows}")
        
    summ = 0
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        outliers = df[(df[column] < (Q1 - 1.5 * (Q3 - Q1)) ) | (df[column] > (Q3 + 1.5 * (Q3 - Q1)) )]
        num_outliers = outliers.shape[0]
        if num_outliers > 0:
            summ += num_outliers
            print(f'Обнаружены выбросы в столбце "{column}":')
            print(outliers[column])
            print()
        
        df.drop(outliers.index, inplace=True) 
        
    
    if summ > 0:
        print(f'Количество выбросов, обнаруженных по методу IQR: {summ}')
    else:
        print('Выбросов не обнаружено.')
    return df
```

## Определение корреляции между параметрами

Корреляционная матрица — это таблица, которая показывает коэффициенты корреляции между различными переменными в наборе данных. 

Если r = -1, это означает, что существует идеальная отрицательная корреляция.

Если r = 0, это означает, что между двумя переменными нет корреляции.

Если r = 1, это означает, что существует идеальная положительная корреляция.

```python
result_matrix = df[['Cost of Living Index', 'Rent Index', 'Cost of Living Plus Rent Index', 'Groceries Index', 'Restaurant Price Index','Local Purchasing Power Index' ]]
matrix = result_matrix.corr()
print(matrix)
```

```python
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt='.4f', square=True, linewidths=.10, cbar_kws={"shrink": .7})
plt.title('Корреляционная матрица')
plt.show()
```

По полученным результатам можно понять, что большинство параметров имеют среднюю или сильную зависимость, что логично для таблицы в которой указаны индексы уровня жизни (питание, жилье и так далее).
Из исследованных атрибутов все имеют корреляцию.
